```python
def set_url(hostname, port, **query):
  real_url = f"https://{hostname}:{port}?"
  for key in query.keys():
    real_url += f"{key}={query[key]}&"

  return real_url

set_url('localhost','8888',utm_source='facebook')
```

```python
# fibonacci sequence

def fibo(n):
  if n==1 or n ==2:
    return 1
  else:
    return fibo(n-1) + fibo(n-2)

fibo(6)
```

```python
# fibonacci sequence with binet's formula
from math import sqrt


def binet_fibo(n):
  sqrt_5 = sqrt(5)
  return int(( ((1+sqrt_5)**n) - ((1-sqrt_5)**n) ) / (2**n*sqrt_5))

binet_fibo(6)
```

```python
# fibonacci sequence with memoization

fibo_dict = dict({(1,1),(2,1)})
def memoization_fibo(n):
  if n in fibo_dict.keys():
    return fibo_dict[n]
  if n==1 or n==2:
    return 1
  else:
    fibo_dict[n] = memoization_fibo(n-1) + memoization_fibo(n-2)
    return fibo_dict[n]

memoization_fibo(5)
```
