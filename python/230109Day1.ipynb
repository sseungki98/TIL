```python
#2

# str = input("Text me > ")
strs = ["FAST","campus12","FastCampus","1107","dynamite"]
for str in strs:
  print("\n",str)
  print(f"{str}는 알파벳 단독입니까? {str.isalpha()}")
  print(f"{str}는 숫자 단독입니까? {str.isdigit()}")
  print(f"{str}는 알파벳과 숫자의 혼용입니까? {not str.isalpha() and not str.isdigit()}")
  print(f"{str}는 모두 소문자입니까? {str.islower()}" )
  print(f"{str}는 모두 대문자입니까? {str.isupper()}")

```
