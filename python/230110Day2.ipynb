```python
# Practice 1

# with List
animals = ['rabbit','cat','dog','aligator','tiger','lion']
animals.append('hippo')
animals.append('koala')
animals.append('bird')
# animals.extends(['','',''])
animals.sort(reverse= True)
print(animals)
even_animals = animals[::2]
print(even_animals)
```

```python
# with Tuple
tuple_animals = ('rabbit','cat','dog','aligator','tiger','lion')
tuple_animals = list(tuple_animals)
tuple_animals.append('hippo')
tuple_animals.append('koala')
tuple_animals.append('bird')
tuple_animals.sort(reverse= True)
tuple_even_animals = tuple(tuple_animals[::2])
# tuple도 range 표현 가능
tuple_animals = tuple(tuple_animals)
# tuple_even_animals = tuple_animals(::2)
print(tuple_animals)
print(tuple_even_animals)
```

```python
# small quiz
A = 'fastcampus'
B = 'shinhan'

# 1 A U B
set_A = set(A)
set_B = set(B)

HAP = set_A | set_B
print("HAP : ",HAP)

# 2 A n B
GYU = set_A & set_B
print("GYU : ",GYU)

# 3 A - B
CHA = set_A - set_B
print("A CHA B : ",CHA)

# 4 A tra B
TRA = set_A ^ set_B
print("TRA : ",TRA)
```

  - 나라가 USA라면, 주마다 다르다.
